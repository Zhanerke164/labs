GENERATOR

```python
def func(a):
    for i in range(1, a): 
        yield i**2 #return
a=int(input())
for n in func(a):
    print(n)
```

```python
def func(a):
    for i in range(0, a):
        if i%2==0:
            yield i
a=int(input())
for n in func(a):
    print(n, end=",")
```

```python
def func(a):
    for i in range(0, a):
        if i%3==0 and i%4==0:
            yield i
a=int(input())
for n in func(a):
    print(n)
```

```python
def func(a, b):
    for i in range(a, b):
        yield i**2
a=int(input())
b=int(input())
for n in func(a, b):
    print(n)
```

```python
def func(a):
    for i in range(a, 0, -1):
        yield i
a=int(input())
for n in func(a):
    print(n)
```
